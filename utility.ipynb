# CDXML -> SMILEs Converter

```python
from rdkit import Chem

'''Extracts smiles form given .cdxml file'''


file_path = ''


# Read CDXML file
with open(file_path, "r", encoding="utf-8") as f:
    cdxml_string = f.read()

all_smiles = []


# Parse CDXML
mols = Chem.MolsFromCDXML(cdxml_string)

if mols:
    for mol in mols:
        if mol:
            smi = Chem.MolToSmiles(mol, canonical=True)

            parts = smi.split('.')
            
            # remove '*' fragments
            filtered = [p for p in parts if p != '*']
            
            all_smiles.extend(filtered)

    print("All smiles combined : ", '.'.join(all_smiles))
else:
    print("Failed to parse molecule")

print("Done")
```

# CDX -> CDXML Converter

```python
from pycdxml import cdxml_converter
import glob, os

# root directory containing cdx files

base = ''

files = glob.glob(os.path.join(base, "*.CDX"))
os.makedirs(base + '/all_cdxml_files', exist_ok=True)
output = base + '/all_cdxml_files'
print(f"Found {len(files)} root files")
counter=0
for f in files:
# cdx to cdxml
    base_name = os.path.basename(f).split('.')[0]
    print(os.path.join(output,f'{base_name}.cdxml'))
    doc = cdxml_converter.read_cdx(f)    
    cdxml_converter.write_cdxml_file(doc, os.path.join(output,f'{base_name}.cdxml'))


print('Done')
```

# Extracts.cdx files from bulk data

```python
import os
import shutil

root_folder = "/Users/rohithgowda/Downloads/2012"
target_folder = "/Users/rohithgowda/Downloads/all_cdx_files"

os.makedirs(target_folder, exist_ok=True)

counter = 0

for root, dirs, files in os.walk(root_folder):
    for file in files:
        if file.lower().endswith(".cdx"):
            source_path = os.path.join(root, file)
            
            # Create unique filename
            counter += 1
            new_name = f"{counter}_{file}"
            dest_path = os.path.join(target_folder, new_name)
            
            shutil.copy2(source_path, dest_path)

print("Done collecting CDX files.")
```
